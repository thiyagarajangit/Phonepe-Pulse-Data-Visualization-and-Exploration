# 1-Data Extraction and Transformation of Aggregated Transactions

```python
path_DataAggTrans = "C:/Thiyagu/GUVI/Project_2_Phonepe_test/data/aggregated/transaction/country/india/state/"
agg_trans_list = os.listdir(path_DataAggTrans)
columnsDataAggTrans = {'State': [], 'Year': [], 'Quarter': [], 'Transaction_type': [], 'Transaction_count': [], 'Transaction_amount': []}
#Trans state wise list
for state in agg_trans_list:
    aggTrans_state = path_DataAggTrans + state + "/"
    aggTrans_year_list = os.listdir(aggTrans_state)
#Trans state & year wise list    
    for year in aggTrans_year_list:
        aggTrans_year = aggTrans_state + year + "/"
        aggTrans_file_list = os.listdir(aggTrans_year)
        
        for file in aggTrans_file_list:
            aggTrans_file = aggTrans_year + file
            #data = open(aggTrans_file, 'r')
            #data_aggTrans = json.load(data)
            
            data = open(aggTrans_file, 'r')
            data_aggTrans = json.load(data)
            
            for i in data_aggTrans['data']['transactionData']:
                name = i['name']
                count = i['paymentInstruments'][0]['count']
                amount = i['paymentInstruments'][0]['amount']
                columnsDataAggTrans['Transaction_type'].append(name)
                columnsDataAggTrans['Transaction_count'].append(count)
                columnsDataAggTrans['Transaction_amount'].append(amount)
                columnsDataAggTrans['State'].append(state)
                columnsDataAggTrans['Year'].append(year)
                columnsDataAggTrans['Quarter'].append(int(file.strip('.json')))           
df_DataAggTrans = pd.DataFrame(columnsDataAggTrans)
```

# 2-Data Extraction and Transformation of Aggregated User

```python
path_DataAggUsers = "C:/Thiyagu/GUVI/Project_2_Phonepe/data/aggregated/user/country/india/state/"
agg_user_list = os.listdir(path_DataAggUsers)
columnsDataAggUsers = {'State': [], 'Year': [], 'Quarter': [], 'Brands': [], 'Count': [],'Percentage': []}
#state wise list
for state in agg_user_list:
    aggUser_state = path_DataAggUsers + state + "/"
    aggUser_year_list = os.listdir(aggUser_state)
    #AggUsers state & year
    for year in aggUser_year_list:
        aggUser_year = aggUser_state + year + "/"
        aggUser_file_list = os.listdir(aggUser_year)

        for file in aggUser_file_list:
            aggUser_file = aggUser_year + file
            data = open(aggUser_file, 'r')
            data_aggUser = json.load(data)
            try:
                for j in data_aggUser["data"]["usersByDevice"]:
                    brand_name = j["brand"]
                    counts = j["count"]
                    percents = j["percentage"]
                    columnsDataAggUsers["Brands"].append(brand_name)
                    columnsDataAggUsers["Count"].append(counts)
                    columnsDataAggUsers["Percentage"].append(percents)
                    columnsDataAggUsers["State"].append(state)
                    columnsDataAggUsers["Year"].append(year)
                    columnsDataAggUsers["Quarter"].append(int(file.strip('.json')))
            except:
                pass
df_DataAggUser = pd.DataFrame(columnsDataAggUsers)
```

## 3-Data Extraction and Transformation of Map Transactions

```python
path_DataMapTrans = "C:/Thiyagu/GUVI/Project_2_Phonepe/data/map/transaction/hover/country/india/state/"
map_trans_list = os.listdir(path_DataMapTrans)
columnsDataMapTrans = {'State': [], 'Year': [], 'Quarter': [], 'District': [], 'Count': [], 'Amount': []}
#state wise list
for state in map_trans_list:
    maptrans_state = path_DataMapTrans + state + "/"
    maptrans_year_list = os.listdir(maptrans_state)
    #Map Trans state & year list  
    for year in maptrans_year_list:
        maptrans_year = maptrans_state + year + "/"
        maptrans_file_list = os.listdir(maptrans_year)
     #Map Trans state & year based files 
        for file in maptrans_file_list:
            maptrans_file = maptrans_year + file
            data = open(maptrans_file, 'r')
            data_mapTrans = json.load(data)
            
            for k in data_mapTrans["data"]["hoverDataList"]:
                district = k["name"]
                count = k["metric"][0]["count"]
                amount = k["metric"][0]["amount"]
                columnsDataMapTrans["District"].append(district)
                columnsDataMapTrans["Count"].append(count)
                columnsDataMapTrans["Amount"].append(amount)
                columnsDataMapTrans['State'].append(state)
                columnsDataMapTrans['Year'].append(year)
                columnsDataMapTrans['Quarter'].append(int(file.strip('.json')))
df_DataMapTrans = pd.DataFrame(columnsDataMapTrans)                
```

# 4-Data Extraction and Transformation of Map User

```python
path_DataMapUser = "C:/Thiyagu/GUVI/Project_2_Phonepe/data/map/user/hover/country/india/state/"
map_user_list = os.listdir(path_DataMapUser)
columnsDataMapUser = {"State": [], "Year": [], "Quarter": [], "District": [], "RegisteredUser": [], "AppOpens": []}
#state wise list
for state in map_user_list:
    mapuser_state = path_DataMapUser + state + "/"
    mapuser_year_list = os.listdir(mapuser_state)
    #Map User state & year list  
    for year in mapuser_year_list:
        mapuser_year = mapuser_state + year + "/"
        mapuser_file_list = os.listdir(mapuser_year)
        #Map user state & year based files 
        for file in mapuser_file_list:
            mapuser_file = mapuser_year + file
            data = open(mapuser_file, 'r')
            data_mapUser = json.load(data)
            
            for n in data_mapUser["data"]["hoverData"].items():
                district = n[0]
                registereduser = n[1]["registeredUsers"]
                appOpens = n[1]['appOpens']
                columnsDataMapUser["District"].append(district)
                columnsDataMapUser["RegisteredUser"].append(registereduser)
                columnsDataMapUser["AppOpens"].append(appOpens)
                columnsDataMapUser['State'].append(state)
                columnsDataMapUser['Year'].append(year)
                columnsDataMapUser['Quarter'].append(int(file.strip('.json')))
                
df_DataMapUser = pd.DataFrame(columnsDataMapUser)
```

# 5 : Top Transactions - Data Extraction and Transformation

```python
path_TopTrans = "C:/Thiyagu/GUVI/Project_2_Phonepe/data/top/transaction/country/india/state/"
top_trans_list = os.listdir(path_TopTrans)
columnsTopTrans = {'State': [], 'Year': [], 'Quarter': [], 'Pincode': [], 'Transaction_count': [],'Transaction_amount': []}
#state wise list       
for state in top_trans_list:
    topTrans_state = path_TopTrans + state + "/"
    topTrans_year_list = os.listdir(topTrans_state)
    #Top Trans - state & year list
    for year in topTrans_year_list:
        topTrans_year = topTrans_state + year + "/"
        topTrans_file_list = os.listdir(topTrans_year)       
        #Top Trans - state & year based files 
        for file in topTrans_file_list:
            topTrans_file = topTrans_year + file
            data = open(topTrans_file, 'r')
            dataTopTrans = json.load(data)
            
            for x in dataTopTrans['data']['pincodes']:
                name = x['entityName']
                count = x['metric']['count']
                amount = x['metric']['amount']
                columnsTopTrans['Pincode'].append(name)
                columnsTopTrans['Transaction_count'].append(count)
                columnsTopTrans['Transaction_amount'].append(amount)
                columnsTopTrans['State'].append(state)
                columnsTopTrans['Year'].append(year)
                columnsTopTrans['Quarter'].append(int(file.strip('.json')))
df_DataTopTrans = pd.DataFrame(columnsTopTrans)
```

# 6 : Top User - Data Extraction and Transformation

```python
path_TopUser = "C:/Thiyagu/GUVI/Project_2_Phonepe/data/top/user/country/india/state/"
top_user_list = os.listdir(path_TopUser)
columnsTopUser = {'State': [], 'Year': [], 'Quarter': [], 'Pincode': [], 'RegisteredUsers': []}
#state wise list   
for state in top_user_list:
    topuser_state = path_TopUser + state + "/"
    topuser_year_list = os.listdir(topuser_state)
    #Top User - state & year list
    for year in topuser_year_list:
        topuser_year = topuser_state + year + "/"
        topuser_file_list = os.listdir(topuser_year)
        #Top User - state & year based files 
        for file in topuser_file_list:
            topuser_file = topuser_year + file
            data = open(topuser_file, 'r')
            dataTopUser = json.load(data)
            
            for m in dataTopUser['data']['pincodes']:
                name = m['name']
                registeredUsers = m['registeredUsers']
                columnsTopUser['Pincode'].append(name)
                columnsTopUser['RegisteredUsers'].append(registeredUsers)
                columnsTopUser['State'].append(state)
                columnsTopUser['Year'].append(year)
                columnsTopUser['Quarter'].append(int(file.strip('.json')))
df_DataTopUser = pd.DataFrame(columnsTopUser)
```
